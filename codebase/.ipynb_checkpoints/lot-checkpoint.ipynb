```python
#std ìƒëŒ€ì  ë¹„êµ
std = lot_data_temp['VALUE', 'std']
user_defined_std_rela_criterion = 30
std_criterion = np.percentile(std, user_defined_std_rela_criterion)
std_criterion

mean = lot_data_temp['VALUE', 'mean']
mean_BOB_RELA = 70
mean_WOW_RELA = 30
BOB_mean = np.percentile(mean, mean_BOB_RELA)
WOW_mean = np.percentile(mean, mean_WOW_RELA)
BOB_mean, WOW_mean
```

```python
#BOB WOW ë„ì¶œ
BOB_lots = lot_data_temp.loc[(lot_data_temp.VALUE['std'] < std_criterion) & (lot_data_temp.VALUE['mean'] > BOB_mean)].index
WOW_lots = lot_data_temp.loc[(lot_data_temp.VALUE['std'] < std_criterion) & (lot_data_temp.VALUE['mean'] < WOW_mean)].index
BOB_lots, WOW_lots
lot_BOB_WOW_Group = lot_data.copy()
lot_BOB_WOW_Group['BOB'] = lot_BOB_WOW_Group['ROOT_LOT_ID'].isin(BOB_lots)
lot_BOB_WOW_Group['WOW'] = lot_BOB_WOW_Group['ROOT_LOT_ID'].isin(WOW_lots)
lot_BOB_WOW_Group
```

```python
#ê° ì„¤ë¹„ë³„ ì„±ê³¼ ì •ë³´ ë„ì¶œ
grouped = lot_BOB_WOW_Group.groupby('S_EQP_ID')
eqp_perf = grouped.agg([np.count_nonzero, np.mean, np.std])

BOB = lot_BOB_WOW_Group.loc[lot_BOB_WOW_Group['BOB'] == True].groupby('S_EQP_ID').VALUE.agg(['mean','std'])
BOB.columns = pd.MultiIndex.from_product([['BOB_stat'], BOB.columns])

WOW = lot_BOB_WOW_Group.loc[lot_BOB_WOW_Group['WOW'] == True].groupby('S_EQP_ID').VALUE.agg(['mean','std'])
WOW.columns = pd.MultiIndex.from_product([['WOW_stat'], WOW.columns])


eqp_perf = pd.merge(eqp_perf, BOB, left_index=True, right_index=True, how='left')
eqp_perf = pd.merge(eqp_perf, WOW, left_index=True, right_index=True, how='left')
#eqp_perf.to_csv('/Users/i/Desktop/LAB/SAMSUNG_PROJECT/IMPLE/imple/result/eqp_perf.csv', sep=',')
eqp_perf

eqp_perf['BOB/step'] = 0.0
def BOB_step(eqp):
    count = eqp_perf.loc[eqp]['BOB', 'count_nonzero']
    step = eqp[:8]
    return count/BOB_step_perf[step]

def WOW_step(eqp):
    count = eqp_perf.loc[eqp]['WOW', 'count_nonzero']
    step = eqp[:8]
    return count/WOW_step_perf[step]

eqp_perf['BOB/step'] = eqp_perf.index.to_series().apply(BOB_step)
eqp_perf['WOW/step'] = eqp_perf.index.to_series().apply(WOW_step)
#eqp_perf = eqp_perf.loc[eqp_perf['BIN_VALUE', 'count_nonzero']>user_defined_eqp_size]
eqp_perf
```

```python
#ks test
eqp_stat = pd.concat([eqp_perf['VALUE', 'count_nonzero'], eqp_perf['VALUE', 'mean']], axis=1, keys=['count', 'mean'])

user_defined_alpha_normality = 0.05

def eval_normality(unique_eqp):
    eqp = lot_data[lot_data['S_EQP_ID'] == unique_eqp]
    if len(eqp['VALUE']) > 3:
        d, p_value = stats.shapiro(eqp['VALUE'])
    else:
        p_value = 0.0
    return p_value

eqp_stat['normality_p_value'] = eqp_stat.index.to_series().apply(eval_normality)
eqp_stat['Normality'] = eqp_stat['normality_p_value'] > user_defined_alpha_normality 
eqp_stat
```

```python
#ANOVA

#stepë³„ ì •ë¦¬
unique_Steps = lot_data['STEP_SEQ'].unique()
user_defined_alpha_ANOVA = 0.05

for unique_Step in unique_Steps:
    #lot_dataë¥¼ íŠ¹ì • STEPìœ¼ë¡œ filtering
    filtered = lot_data[lot_data['STEP_SEQ'] == unique_Step]
    
    unique_eqps = lot_data[lot_data['STEP_SEQ'] == unique_Step]['S_EQP_ID'].unique()

    #step ë‚´ ì„¤ë¹„ë“¤ì˜ BIN_VALUEë¥¼ ì €ìž¥
    step_eqp_values = {}
    for unique_eqp in unique_eqps:
        #filtered['S_EQP_ID']ê°€ í•„ìš”, ëª¨ë“  observationì´ í•„ìš”í•˜ë¯€ë¡œ
        step_eqp_values[unique_eqp]=(filtered.loc[filtered['S_EQP_ID']==unique_eqp, 'VALUE'])

    statistic, p_value = stats.f_oneway(*step_eqp_values.values())


    eqp_stat.loc[unique_eqps, 'anova_p_value'] = p_value

eqp_stat['ANOVA'] = eqp_stat['anova_p_value'] < user_defined_alpha_ANOVA
eqp_stat
```

```python

#KRUSKAL
user_defined_alpha_KRUSKAL =0.05
#stepë³„ ì •ë¦¬
unique_Steps = lot_data['STEP_SEQ'].unique()
for unique_Step in unique_Steps:
    #lot_dataë¥¼ íŠ¹ì • STEPìœ¼ë¡œ filtering
    filtered = lot_data[lot_data['STEP_SEQ'] == unique_Step]
    
    unique_eqps = lot_data[lot_data['STEP_SEQ'] == unique_Step]['S_EQP_ID'].unique()

    #step ë‚´ ì„¤ë¹„ë“¤ì˜ BIN_VALUEë¥¼ ì €ìž¥
    step_eqp_values = {}
    for unique_eqp in unique_eqps:
        #filtered['S_EQP_ID']ê°€ í•„ìš”, ëª¨ë“  observationì´ í•„ìš”í•˜ë¯€ë¡œ 
        step_eqp_values[unique_eqp]=(filtered.loc[filtered['S_EQP_ID']==unique_eqp, 'VALUE'])

    statistic, p_value = stats.kruskal(*step_eqp_values.values())


    eqp_stat.loc[unique_eqps, 'kruskal_p_value'] = p_value
eqp_stat['KRUSKAL'] = eqp_stat['kruskal_p_value'] < user_defined_alpha_KRUSKAL
eqp_stat
```

```python
def Linear_Contrast(row, other_rows, unique_Step):
    #N is total observations
    N = sum(other_rows['VALUE', 'count_nonzero'])
    
    filtered = lot_data[lot_data['STEP_SEQ'] == unique_Step]
    filtered_VALUE = filtered.loc[filtered['S_EQP_ID'].isin(other_rows.index), 'VALUE']
    
    #calculate SS_error
    sum_y_squared = sum([value**2 for value in filtered_VALUE])
    SS_error = sum_y_squared - sum(filtered.groupby('S_EQP_ID').sum()['VALUE']**2/filtered.groupby('S_EQP_ID').size())
    
    #exclude ë¹„êµëŒ€ìƒ
    other_rows = other_rows.drop(row.name)
    
    #k is # groups
    k = len(other_rows.index)
    
    #calcualte SS contrast
    L = k*row['VALUE', 'mean']
    for mean in other_rows['VALUE', 'mean']:
        L -= mean
    
    denominator = k*k/row['VALUE', 'count_nonzero']
    for no_obs in other_rows['VALUE', 'count_nonzero']:
        denominator += 1/no_obs
        N +=no_obs
        
    #degrees of freedom
    DFbetween = 1
    DFwithin = N - k
    
    MS_error = SS_error / DFwithin
    
    std = math.sqrt(MS_error*denominator)
    
    t_contrast = abs(L)/std
    
    p_value = stats.t.sf(t_contrast, DFwithin)
    return p_value

def Nonparametric_Linear_Contrast(row, other_rows, unique_Step):
    #N is total observations
    other_rows = other_rows.drop(row.name)
    
    
    filtered = lot_data[lot_data['STEP_SEQ'] == unique_Step]
    other_rows_filtered_VALUE = filtered.loc[filtered['S_EQP_ID'].isin(other_rows.index), 'VALUE']
    #print(other_rows_filtered_VALUE)
    row_filtered_VALUE = filtered.loc[filtered['S_EQP_ID'] == row.name, 'VALUE']
    #print(row_filtered_VALUE)
    statistic, p_value = stats.mannwhitneyu(other_rows_filtered_VALUE, row_filtered_VALUE)
    #print(p_value)
    return p_value
```

```python
#í†µê³„ë¶„ì„
#ê¸°ë³¸ì ìœ¼ë¡œ DUMMYí™” ì‹œì¼œë†“ê³ , BOB/WOW ì„¤ë¹„ë§Œ ìž¬ì„¤ì •

eqp_stat_LC = eqp_stat.copy()
user_defined_alpha_LC = 0.01
user_defined_alpha_NLC = 0.01

eqp_stat_LC['LC_p_value'] = 1.0
eqp_stat_LC['LCresult'] = 'DUMMY'

unique_Steps = lot_data['STEP_SEQ'].unique()
#count_k=0
#count_a=0
#count = 0
for unique_Step in unique_Steps:
    #count +=1
    #lot_dataë¥¼ íŠ¹ì • STEPìœ¼ë¡œ filtering
    filtered = lot_data[lot_data['STEP_SEQ'] == unique_Step]
    #í•´ë‹¹ stepì˜ eqp_stat, filtered['S_EQP_ID'].unique()ê°€ í•„ìš”, ì–´ë–¤ eqpê°€ ìžˆëŠ”ì§€ê°€ í•„ìš”í•˜ë¯€ë¡œ
    step_eqp_stat = eqp_stat_LC.loc[filtered['S_EQP_ID'].unique()]
    
    #pass Normality test
    if all(step_eqp_stat['normality_p_value'] > user_defined_alpha_normality):
        #count_a+=1
        #ANOVA result: ìœ ì˜ë¯¸í•œ ì°¨ì´ ì—†ìŒ
        if (step_eqp_stat['anova_p_value'] > user_defined_alpha_ANOVA).any():
            #eqp_stat.loc[step_eqp_stat.index]['result'] = 'DUMMY' #ì™œ ì•ˆë˜ëƒ?
            eqp_stat_LC['LCresult'][step_eqp_stat.index] = 'DUMMY'
            #print(eqp_stat.loc[step_eqp_stat.index]['result'])
        else:
            #Linear Contrast ìˆ˜í–‰
            sorted_eqp_perf = eqp_perf.loc[filtered['S_EQP_ID'].unique()].sort_values(by=[('VALUE','mean')],ascending=False)
            #print(sorted_eqp_perf)
            for index, row in sorted_eqp_perf.iterrows():
                p_value = Linear_Contrast(row, eqp_perf.loc[filtered['S_EQP_ID'].unique()], unique_Step)
                eqp_stat_LC['LC_p_value'][row.name]  = p_value
                ###Linear Contrast ê²°ê³¼ ìœ ì˜ë¯¸í•œ ê²½ìš°
                if p_value < user_defined_alpha_LC:
                    eqp_stat_LC['LCresult'][row.name] = 'BOB'
                else:
                    break
            sorted_eqp_perf = eqp_perf.loc[filtered['S_EQP_ID'].unique()].sort_values(by=[('VALUE','mean')])            
            for index, row in sorted_eqp_perf.iterrows():
                p_value = Linear_Contrast(row, eqp_perf.loc[filtered['S_EQP_ID'].unique()], unique_Step)
                eqp_stat_LC['LC_p_value'][row.name]  = p_value
                if p_value < user_defined_alpha_LC:
                    eqp_stat_LC['LCresult'][row.name] = 'WOW'
                else:
                    break
                    
    #Non-parametric method
    else:
        #count_k+=1
        if (step_eqp_stat['kruskal_p_value'] > user_defined_alpha_KRUSKAL).any():
            eqp_stat_LC['LCresult'][step_eqp_stat.index] = 'DUMMY'
            
        else:
            ###non-parametric linear contrast
            
            #from the best
            sorted_eqp_perf = eqp_perf.loc[filtered['S_EQP_ID'].unique()].sort_values(by=[('VALUE','mean')],ascending=False)
            for index, row in sorted_eqp_perf.iterrows():
                p_value = Nonparametric_Linear_Contrast(row, eqp_perf.loc[filtered['S_EQP_ID'].unique()], unique_Step)
                eqp_stat_LC['LC_p_value'][row.name]  = p_value
                ###Linear Contrast ê²°ê³¼ ìœ ì˜ë¯¸í•œ ê²½ìš°
                if p_value < user_defined_alpha_NLC:
                    eqp_stat_LC['LCresult'][row.name] = 'BOB'
                else:
                    eqp_stat_LC['LCresult'][row.name] = 'DUMMY'
                    break
            #from the worst
            sorted_eqp_perf = eqp_perf.loc[filtered['S_EQP_ID'].unique()].sort_values(by=[('VALUE','mean')])            
            for index, row in sorted_eqp_perf.iterrows():
                p_value = Nonparametric_Linear_Contrast(row, eqp_perf.loc[filtered['S_EQP_ID'].unique()], unique_Step)
                eqp_stat_LC['LC_p_value'][row.name]  = p_value
                if p_value < user_defined_alpha_NLC:
                    eqp_stat_LC['LCresult'][row.name] = 'WOW'
                else:
                    eqp_stat_LC['LCresult'][row.name] = 'DUMMY'
                    break
    
#WOW í™•ì¸
#eqp_stat_LC.loc[eqp_stat_LC['LCresult']=='WOW']  
eqp_stat_LC
                        
                        
                        
```

```python
#ì‚¬ìš©ìž ì§€ì • ìˆ˜ìœ¨ ê¸°ì¤€ ë¶„ì„
eqp_perf_user_defined = eqp_perf.copy()

user_defined_BOB_avg = 8
user_defined_WOW_avg = 7

user_defined_BOB_Prop = 0.20
user_defined_WOW_Prop = 0.20

user_defined_eqp_weight = 0.5
user_defined_step_weight = 0.5


eqp_perf_user_defined['AVG_Test'] = np.where(eqp_perf_user_defined['VALUE', 'mean']  > user_defined_BOB_avg, 'BOB', np.where(eqp_perf_user_defined['VALUE', 'mean'] < user_defined_WOW_avg, 'WOW', 'DUMMY'))
eqp_perf_user_defined['Prop_Test'] = np.where(user_defined_eqp_weight*eqp_perf_user_defined['BOB', 'mean'] + user_defined_step_weight*eqp_perf_user_defined['BOB/step'] > user_defined_BOB_Prop, np.where(user_defined_eqp_weight*eqp_perf_user_defined['WOW', 'mean'] + user_defined_step_weight*eqp_perf_user_defined['WOW/step'] > user_defined_WOW_Prop, 'DUMMY', 'BOB'), np.where(user_defined_eqp_weight*eqp_perf_user_defined['WOW', 'mean'] + user_defined_step_weight*eqp_perf_user_defined['WOW/step'] > user_defined_WOW_Prop, 'WOW', 'DUMMY'))
eqp_perf_user_defined
```

```python
mode = 'and'
mode = 'or'

def and_assign(eqp):
    x = eqp_perf_user_defined.loc[eqp]['AVG_Test']
    y = eqp_perf_user_defined.loc[eqp]['Prop_Test']
    x = x.values
    y = y.values
    if x==y:
        return x[0]
    else:
        return 'DUMMY'

def or_assign(eqp):
    x = eqp_perf_user_defined.loc[eqp]['AVG_Test']
    y = eqp_perf_user_defined.loc[eqp]['Prop_Test']
    x = x.values
    y = y.values
    x = x[0]
    y = y[0]
    if x==y:
        return x
    else:
        if x != 'DUMMY':
            return x
        elif y != 'DUMMY':
            return y
        else:
            return 'DUMMY'
        

if mode == 'and':
    eqp_perf_user_defined['UDresult'] = eqp_perf_user_defined.index.to_series().apply(and_assign)
elif mode == 'or':
    eqp_perf_user_defined['UDresult'] = eqp_perf_user_defined.index.to_series().apply(or_assign)
eqp_perf_user_defined
```
